**Practical 3. Implement an N-gram Language Model in Python to analyze a text sequence and predict the most probable next word**

**What is an N-gram?**

An N-gram is a contiguous sequence of N items (words) from a given text.

- **Unigram (N=1)** -> Single word: ['the'], ['dog'], ['runs']
- **Bigram (N=2)** -> Two consecutive words: ['the dog'], ['dog runs']
- **Trigram (N=3)** -> Three consecutive words: ['the dog runs'], ['dog runs fast']

**How does it predict the next word?**

It looks at the previous (N-1) words and finds the most frequent word that followed those words in the training corpus.

1. **Download required NLTK packages and import libraries**

```python
import nltk
from nltk.util import ngrams
from collections import Counter, defaultdict

nltk.download('punkt_tab')
```

2. **Define a small text corpus for training the N-gram model**

```python
corpus = """
the cat sat on the mat the cat ate the rat the rat ran away
the dog sat on the log the dog barked at the cat the cat ran away
the bird flew over the tree the bird sat on the branch the branch fell down
the sun rises in the morning the sun sets in the evening the stars shine at night
the student reads the book the student writes the answer the teacher checks the answer
"""
```

3. **Tokenize the corpus into words (lowercase)**

```python
tokens = nltk.word_tokenize(corpus.lower())
print("Total Tokens:", len(tokens))
print()
print(tokens)
```

4. **Generate Unigrams (N=1) and display their frequency**

```python
unigrams = list(ngrams(tokens, 1))
unigram_freq = Counter(unigrams)

print("Unigrams (Top 10):")
for gram, count in unigram_freq.most_common(10):
    print(gram, " : ", count)
```

5. **Generate Bigrams (N=2) and display their frequency**

```python
bigrams = list(ngrams(tokens, 2))
bigram_freq = Counter(bigrams)

print("Bigrams (Top 10):")
for gram, count in bigram_freq.most_common(10):
    print(gram, " : ", count)
```

6. **Generate Trigrams (N=3) and display their frequency**

```python
trigrams = list(ngrams(tokens, 3))
trigram_freq = Counter(trigrams)

print("Trigrams (Top 10):")
for gram, count in trigram_freq.most_common(10):
    print(gram, " : ", count)
```

7. **Build a Bigram Language Model using defaultdict**

Logic:
- For every bigram (word1, word2), store word2 as a possible next word after word1
- Count how many times each next word appeared
- This gives us a probability table: given word1 -> what is the most likely word2?

```python
bigram_model = defaultdict(Counter)

for w1, w2 in bigrams:
    bigram_model[w1][w2] += 1

print("Bigram Model (showing next word counts for some words):")
print()
for word in ['the', 'cat', 'sat', 'student']:
    print(f"After '{word}' -> ", dict(bigram_model[word]))
```

8. **Build Bigram Next Word Predictor Function**

Method -> predict_next_word(word) returns the most probable next word

```python
def predict_next_word_bigram(word):
    word = word.lower()
    if word in bigram_model:
        next_word = bigram_model[word].most_common(1)[0][0]
        return next_word
    else:
        return "Word not found in corpus"
```

```python
test_words = ['the', 'cat', 'sat', 'dog', 'student', 'sun', 'bird']

print("Bigram Next Word Predictions:")
print()
for w in test_words:
    print(w, " : ", predict_next_word_bigram(w))
```

9. **Build a Trigram Language Model for better context-aware prediction**

Logic:
- For every trigram (word1, word2, word3), store word3 as a possible next word after (word1, word2)
- Given the last 2 words, predict the most probable next word

```python
trigram_model = defaultdict(Counter)

for w1, w2, w3 in trigrams:
    trigram_model[(w1, w2)][w3] += 1

print("Trigram Model (showing next word counts for some word pairs):")
print()
for pair in [('the', 'cat'), ('sat', 'on'), ('the', 'sun'), ('the', 'student')]:
    print(f"After {pair} -> ", dict(trigram_model[pair]))
```

10. **Build Trigram Next Word Predictor Function**

Method -> predict_next_word_trigram(word1, word2) returns the most probable next word

```python
def predict_next_word_trigram(word1, word2):
    word1, word2 = word1.lower(), word2.lower()
    if (word1, word2) in trigram_model:
        next_word = trigram_model[(word1, word2)].most_common(1)[0][0]
        return next_word
    else:
        return "Word pair not found in corpus"
```

```python
test_pairs = [
    ('the', 'cat'),
    ('sat', 'on'),
    ('the', 'dog'),
    ('the', 'sun'),
    ('the', 'student'),
    ('the', 'bird'),
    ('cat', 'ran')
]

print("Trigram Next Word Predictions:")
print()
for w1, w2 in test_pairs:
    print(f"('{w1}', '{w2}')  :  ", predict_next_word_trigram(w1, w2))
```

11. **Generate a sentence using the Bigram model (auto text generation)**

Start with a seed word and keep predicting the next word for N steps

```python
def generate_sentence_bigram(start_word, num_words=8):
    sentence = [start_word]
    current_word = start_word.lower()
    for _ in range(num_words - 1):
        next_word = predict_next_word_bigram(current_word)
        if next_word == "Word not found in corpus":
            break
        sentence.append(next_word)
        current_word = next_word
    return ' '.join(sentence)

print("Generated Sentences using Bigram Model:")
print()
for seed in ['the', 'cat', 'dog', 'student']:
    print(f"Seed: '{seed}'  ->  ", generate_sentence_bigram(seed))
```
