**Practical 4. Implement and Train Basic Word Embeddings using a Limited Text Corpus in Python to Investigate Semantic Similarities and Relationships Among Words**

**What are Word Embeddings?**

Word Embeddings are dense vector representations of words in a continuous vector space, where semantically similar words are mapped to nearby points.

**Key Idea:** Words with similar meaning appear in similar contexts, so their vectors will be close to each other.

**Word2Vec Model Types:**
- **CBOW (Continuous Bag of Words)** -> Predicts the center word from surrounding context words
- **Skip-gram** -> Predicts surrounding context words from the center word

**Important Parameters:**
- **vector_size** -> Dimension of the word vector (e.g. 100 means each word = 100 numbers)
- **window** -> Context window size (how many words left/right to consider)
- **min_count** -> Ignore words with frequency less than this value
- **sg** -> 0 = CBOW, 1 = Skip-gram

1. **Install and import required libraries**

```python
import nltk
import warnings
warnings.filterwarnings('ignore')

nltk.download('punkt_tab')
nltk.download('stopwords')

from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import string

print("All libraries imported successfully!")
```

2. **Define a small text corpus for training Word2Vec**

```python
corpus = [
    "the king rules the kingdom with power and authority",
    "the queen rules the kingdom with grace and wisdom",
    "the prince is the son of the king and the queen",
    "the princess is the daughter of the king and the queen",
    "the king and the queen live in the royal palace",
    "the man went to the market to buy fruits and vegetables",
    "the woman went to the market to buy flowers and groceries",
    "the boy plays cricket in the ground every morning",
    "the girl plays tennis in the court every evening",
    "the dog and the cat are domestic animals and loyal pets",
    "the lion and the tiger are wild animals found in the forest",
    "the doctor treats sick patients in the hospital every day",
    "the nurse helps the doctor treat patients in the hospital",
    "the teacher teaches students in the school every morning",
    "the professor teaches students in the university every day",
    "india and china are large countries in asia",
    "france and germany are large countries in europe",
    "paris is the capital city of france in europe",
    "delhi is the capital city of india in asia",
    "the scientist discovered a new planet beyond the solar system",
    "the astronaut travelled to the moon on a powerful rocket",
    "deep learning and machine learning are subfields of artificial intelligence",
    "python is a popular programming language used in machine learning",
    "neural networks are used in deep learning and artificial intelligence"
]
```

3. **Preprocess the corpus -> Tokenize and remove punctuation**

```python
def preprocess(sentence):
    tokens = word_tokenize(sentence.lower())
    tokens = [t for t in tokens if t not in string.punctuation]
    return tokens

tokenized_corpus = [preprocess(sentence) for sentence in corpus]

print("Tokenized Corpus (first 5 sentences):")
print()
for i, tokens in enumerate(tokenized_corpus[:5]):
    print(f"Sentence {i+1}  :  {tokens}")
```

4. **Train the Word2Vec model (CBOW) on the tokenized corpus**

- **vector_size = 100** -> Each word will be represented as a 100-dimensional vector
- **window = 5** -> Look 5 words to the left and right as context
- **min_count = 1** -> Include all words even if they appear only once
- **sg = 0** -> Use CBOW model

```python
model = Word2Vec(
    sentences=tokenized_corpus,
    vector_size=100,
    window=5,
    min_count=1,
    sg=0,
    epochs=200,
    seed=42
)

print("Word2Vec model trained successfully!")
print()
print("Total words in vocabulary  :  ", len(model.wv))
```

5. **Display the vocabulary learned by the model**

```python
vocab = list(model.wv.index_to_key)
print("Vocabulary learned by Word2Vec model:")
print()
print(vocab)
```

6. **Display the Word Vector (embedding) for a specific word**

Method -> **model.wv['word']** returns the vector of that word

```python
print("Word Vector for 'king' (100 dimensions):")
print()
print(model.wv['king'])
```

7. **Find Most Similar Words using Cosine Similarity**

Method -> **model.wv.most_similar('word', topn=5)**

Returns top N words most similar to the given word based on their vector distance

```python
test_words = ['king', 'queen', 'doctor', 'teacher', 'dog', 'france']

print("Most Similar Words (Top 3):")
print()
for word in test_words:
    similar = model.wv.most_similar(word, topn=3)
    print(f"'{word}'  :")
    for sim_word, score in similar:
        print(f"     {sim_word}  :  {round(score, 4)}")
    print()
```

8. **Calculate Cosine Similarity between two specific words**

Method -> **model.wv.similarity('word1', 'word2')**

Returns a value between -1 and 1:
- Score close to **1** -> words are very similar
- Score close to **0** -> words are unrelated
- Score close to **-1** -> words are opposite in context

```python
word_pairs = [
    ('king', 'queen'),
    ('king', 'man'),
    ('doctor', 'nurse'),
    ('teacher', 'professor'),
    ('dog', 'cat'),
    ('india', 'france'),
    ('paris', 'delhi'),
    ('king', 'tiger')
]

print("Cosine Similarity between Word Pairs:")
print()
for w1, w2 in word_pairs:
    score = model.wv.similarity(w1, w2)
    print(f"('{w1}', '{w2}')  :  {round(score, 4)}")
```

9. **Word Analogy: king - man + woman = ?**

Method -> **model.wv.most_similar(positive=['king', 'woman'], negative=['man'], topn=1)**

This is the famous Word2Vec analogy test:
- **king - man + woman** should give **queen**
- **india - delhi + paris** should give **france**

```python
print("Word Analogy Tests:")
print()

# king - man + woman = ?
result = model.wv.most_similar(positive=['king', 'woman'], negative=['man'], topn=1)
print("king - man + woman  :  ", result[0][0], " (score:", round(result[0][1], 4), ")")

# queen - woman + man = ?
result = model.wv.most_similar(positive=['queen', 'man'], negative=['woman'], topn=1)
print("queen - woman + man  :  ", result[0][0], " (score:", round(result[0][1], 4), ")")

# doctor - man + woman = ?
result = model.wv.most_similar(positive=['doctor', 'woman'], negative=['man'], topn=1)
print("doctor - man + woman  :  ", result[0][0], " (score:", round(result[0][1], 4), ")")

# india - delhi + paris = ?
result = model.wv.most_similar(positive=['india', 'paris'], negative=['delhi'], topn=1)
print("india - delhi + paris  :  ", result[0][0], " (score:", round(result[0][1], 4), ")")
```

10. **Find the Odd Word Out from a list of words**

Method -> **model.wv.doesnt_match(['word1', 'word2', 'word3', 'word4'])**

Returns the word that does not fit with the others based on their vector positions

```python
word_lists = [
    ['king', 'queen', 'prince', 'dog'],
    ['doctor', 'nurse', 'hospital', 'tiger'],
    ['india', 'france', 'germany', 'cat'],
    ['teacher', 'professor', 'school', 'lion']
]

print("Odd Word Out:")
print()
for word_list in word_lists:
    odd = model.wv.doesnt_match(word_list)
    print(word_list, "  ->  Odd word:", odd)
```

11. **Train Word2Vec using Skip-gram model (sg=1) and compare**

- **CBOW (sg=0)** -> Faster to train, works better with frequent words
- **Skip-gram (sg=1)** -> Slower but works better with rare words and small corpus

```python
skipgram_model = Word2Vec(
    sentences=tokenized_corpus,
    vector_size=100,
    window=5,
    min_count=1,
    sg=1,
    epochs=200,
    seed=42
)

print("Skip-gram model trained successfully!")
print()
print("Comparing CBOW vs Skip-gram for word 'king' (Top 3 similar words):")
print()

cbow_similar = model.wv.most_similar('king', topn=3)
skip_similar = skipgram_model.wv.most_similar('king', topn=3)

print("CBOW:")
for word, score in cbow_similar:
    print("    ", word, " : ", round(score, 4))

print()
print("Skip-gram:")
for word, score in skip_similar:
    print("    ", word, " : ", round(score, 4))
```

12. **Save and Load the trained Word2Vec model**

```python
# Save the model
model.save("word2vec_cbow.model")
print("Model saved as 'word2vec_cbow.model'")

# Load the model
loaded_model = Word2Vec.load("word2vec_cbow.model")
print("Model loaded successfully!")
print()

# Verify it works
print("Verification - Most similar to 'queen' from loaded model:")
result = loaded_model.wv.most_similar('queen', topn=3)
for word, score in result:
    print("    ", word, " : ", round(score, 4))
```
