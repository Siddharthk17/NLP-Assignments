**Practical 2. Develop a Python script to perform Part-of-Speech (POS) Tagging on a short text dataset**

**What is POS Tagging?**

Part-of-Speech (POS) Tagging is the process of assigning a grammatical label (like Noun, Verb, Adjective, etc.) to each word in a sentence.

**Common POS Tags:**
- **NN** -> Noun (singular) e.g. dog, city
- **NNS** -> Noun (plural) e.g. dogs, cities
- **NNP** -> Proper Noun (singular) e.g. NASA, India
- **VB** -> Verb (base form) e.g. run, eat
- **VBD** -> Verb (past tense) e.g. ran, ate
- **VBG** -> Verb (gerund/present participle) e.g. running, eating
- **JJ** -> Adjective e.g. big, fast
- **RB** -> Adverb e.g. quickly, very
- **DT** -> Determiner e.g. the, a, an
- **IN** -> Preposition e.g. in, on, at
- **CC** -> Coordinating Conjunction e.g. and, but, or
- **PRP** -> Personal Pronoun e.g. he, she, they

1. **Download required NLTK packages: punkt_tab and averaged_perceptron_tagger**

```python
import nltk
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
```

2. **Define a small text corpus (dataset of 3 sentences)**

```python
corpus = [
    "The Indian Space Research Organisation successfully launched a new satellite into orbit.",
    "Doctors are working tirelessly to develop an effective vaccine against the deadly virus.",
    "The young students quickly solved the difficult mathematics problem in the examination hall."
]
```

3. **Tokenize each sentence and display tokens**

```python
for i, sentence in enumerate(corpus):
    tokens = nltk.word_tokenize(sentence)
    print(f"Sentence {i+1} Tokens:")
    print(tokens)
    print()
```

4. **Perform POS Tagging using nltk.pos_tag() on each sentence**

Method -> **nltk.pos_tag(tokens)**

Returns a list of tuples: **(word, POS_tag)**

```python
for i, sentence in enumerate(corpus):
    tokens = nltk.word_tokenize(sentence)
    pos_tags = nltk.pos_tag(tokens)
    print(f"Sentence {i+1} POS Tags:")
    print(pos_tags)
    print()
```

5. **Display POS Tags in a clean Word : Tag format (like Stemming/Lemmatization output)**

```python
for i, sentence in enumerate(corpus):
    tokens = nltk.word_tokenize(sentence)
    pos_tags = nltk.pos_tag(tokens)
    print(f"Sentence {i+1}")
    for word, tag in pos_tags:
        print(word, " : ", tag)
    print()
```

6. **Filter and display only Nouns (NN, NNS, NNP) from all sentences**

```python
noun_tags = ['NN', 'NNS', 'NNP', 'NNPS']

for i, sentence in enumerate(corpus):
    tokens = nltk.word_tokenize(sentence)
    pos_tags = nltk.pos_tag(tokens)
    nouns = [word for word, tag in pos_tags if tag in noun_tags]
    print(f"Sentence {i+1} Nouns: {nouns}")
```

7. **Filter and display only Verbs (VB, VBD, VBG, VBP) from all sentences**

```python
verb_tags = ['VB', 'VBD', 'VBG', 'VBN', 'VBP', 'VBZ']

for i, sentence in enumerate(corpus):
    tokens = nltk.word_tokenize(sentence)
    pos_tags = nltk.pos_tag(tokens)
    verbs = [word for word, tag in pos_tags if tag in verb_tags]
    print(f"Sentence {i+1} Verbs: {verbs}")
```

8. **Filter and display only Adjectives (JJ, JJR, JJS) from all sentences**

```python
adjective_tags = ['JJ', 'JJR', 'JJS']

for i, sentence in enumerate(corpus):
    tokens = nltk.word_tokenize(sentence)
    pos_tags = nltk.pos_tag(tokens)
    adjectives = [word for word, tag in pos_tags if tag in adjective_tags]
    print(f"Sentence {i+1} Adjectives: {adjectives}")
```

9. **Summary: Count of each POS tag across the entire corpus**

```python
from collections import Counter

all_tags = []

for sentence in corpus:
    tokens = nltk.word_tokenize(sentence)
    pos_tags = nltk.pos_tag(tokens)
    for word, tag in pos_tags:
        all_tags.append(tag)

tag_counts = Counter(all_tags)

print("POS Tag Frequency across entire corpus:")
print()
for tag, count in tag_counts.most_common():
    print(tag, " : ", count)
```
