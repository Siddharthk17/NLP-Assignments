**Practical 1. Implement tokenization, stemming, and lemmatization on a small corpus of text**

**Paragraph -> Corpus**
**Sentence -> Document**
**Unique Words -> Vocabulary**

1. **Tokenization -> punkt_tab**
  From Sentence to Words -> word_tokenize()
  From Corpus to Sentence -> sent_tokenize()

```python
import nltk
nltk.download("punkt_tab")
```

```python
sentence = """Scientists at NASA have announced the discovery of water ice beneath the surface of Mars, suggesting that the red planet may have once supported liquid water and potentially microbial life. The findings, published in a leading scientific journal, were based on data collected by the Mars Reconnaissance Orbiter over several years of observation."""
```

```python
tokens = nltk.word_tokenize(sentence)
```

```python
tokens
```

```python
sentence
```

```python
nltk.sent_tokenize(sentence)
```

**2. Stemming -> Finding Root Word using PorterStemmer**

**Method -> ps.stem(each word in tokens array)**

Types:
Porter Stemmer, Snowball Stemmer, Regexp Stemmer, Lancaster Stemmer

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
for t in tokens:
  print(t, " : ", ps.stem(t))
```

**3. Lemmatization: Download wordnet from nltk and use WordNetLemmatizer**

**Method: lemma.lemmatize(each word in tokens array)**

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
lemma = WordNetLemmatizer()
```

```python
for t in tokens:
  print(t, " : ", lemma.lemmatize(t))
```

4. Identify & Remove Stopwords
Download stopwords from nltk
Method -> stopwords.words(Language)

```python
from nltk.corpus import stopwords
nltk.download("stopwords")

sw = set(stopwords.words("english"))

filtered_tokens = [t for t in tokens if t.lower() not in sw]

print(filtered_tokens)
```
